# Introduction

### Le neurone

Un neurone artificiel (ou neurone formel - https://fr.wikipedia.org/wiki/Neurone_formel) reproduit de façon simplifiée le fonctionnement d'un neurone biologique. Les deux sont comparés dans l'image ci-dessous:

<img src="./NeuroneArtificiel/neurones_schemas.png" alt="" style="width: 700px;"/>

<div align="center">Fig. 1: Comparaison d'un neurone biologique et d'un neurone artificiel.</div>

Le neurone biologique est stimulé par les dendrites. Celles-ci reçoivent des signaux de stimulation de diverses manières, par exemple par d'autres neurones ou par des cellules sensorielles. Le corps cellulaire fait la somme de ces stimulations. Si la somme des stimulations des dendrites est supérieure à un certain seuil (appelé le 'seuil d'excitabilité' du neurone), le neurone s'excite: Il transmet des impulsions électriques à travers son axone, ce qui amène à stimuler les dendrites des neurones qui y sont connectés. 

Le neurone artificiel simule ce fonctionnement. Les entrées du neurone artificiel joue le rôle des dendrites, et la sortie celui de l'axone. Le corps cellulaire est représenté par un rond. Le symbole '+' est placé au centre du rond pour indiquer que la somme des entrées y est effectuée. Si cette somme est supérieure au seuil d'excitabilité appelé T, alors le neurone s'excite et la sortie est à 1. Sinon, la sortie reste à 0. 

Exemple: Supposons que l'entrées 1 envoie une valeur de 3, l'entrée 3 envoie une valeur de -1, et les autes entrées n'envoient aucun signal (zéro). La somme des entrées est donc de 2. Si le seuil d'excitabilité du neurone est de 1, alors le neurone émettra un signal en sortie de 1 (somme des entrées>seuil d'excitabilité: 2>1). 

### Communication entre neurones: Les synapses

Les neurones communiquent entre eux grâce aux *synapses*. Une synapse est un petit espace entre un axone terminal et une dendrite. 

<img src="NeuroneArtificiel/synapse.png" alt="" style="width: 500px;"/>

<div align="center">Fig. 2: Synapse.</div>

La communication se fait de façon chimique par l'intermédiaire de messagers chimiques: les neurotransmetteurs. 

* Voir: https://www.youtube.com/watch?v=yY_v7e4xnf8



### Apprentissage: La règle de Hebb

Les liens que forment les neurones dans le cerveau changent au cours du temps. C'est ce que l'on appelle la plasticité cérébrale, et qui nous permet par exemple d'apprendre (ou d'oublier). Le phénomène d'apprentissage dans le cerveau est complexe, et est loin d'être précisément compris. C'est l'objet d'étude des neurosciences.

* Vidéo de cellules du cerveau créant de nouvelles connexions: https://www.youtube.com/watch?v=RxBQ1ICEnKE

L'étude du cerveau a permis de montrer que les neurones tendent à se connecter lorsqu'ils sont excités au même moment. Par exemple, les aliments que nous aimons bien nous font saliver, et ceux que nous n'aimons pas provoquent du dégoût. Au fur et à mesure que nous goûtons différents aliments ou recettes, notre cerveau apprend les combinaisons de ceux qui nous plaisent ou nous déplaisent. 

Ce principe est à la base de l'apprentissage dans le cerveau, et pour les neurones artificiels. Il a été proposé en 1949 par Donald Hebb, et s'appelle aussi la [*règle de Hebb*](https://fr.wikipedia.org/wiki/R%C3%A8gle_de_Hebb). La régle se résume par "des neurones qui s'excitent ensemble se lient entre eux". 

Les neurones peuvent se stimuler positivement (de bons aliments amènent à avoir envie de manger), ou négativement (les choses dégoûtantes nous coupent l'appétit). Dans les neurones biologiques, la stimulation positive (excitatrice) ou négative (inhibitrice) est créée par le type de neurotransmetteurs qu'un neurone libérera dans une synapse. Ceci est simulé dans un neurone artificiel par la transmission de valeurs positives ou négatives. Dans l'activité ci-dessous, nous utiliserons des jetons pour représenter ces valeurs: des jetons verts, dont le nombre représentera l'importance de l'excitation, ou des jetons rouges, dont le nombre représentera l'importance de l'inhibitation. 


# Activité 1: Le neurone 'miam' d'une bonne pizza


## Le neurone de la bonne pizza

Dans cette première activité, nous allons jouer le rôle d'un neurone artificiel qui détecte les ingrédients d'une bonne pizza. Nous l'appelerons le neurone 'miam'. 

Ce neurone reçoit en entrée les stimulations de neurones reconnaissant les ingrédients dont peuvent être constituée la pizza. Nous supposons que 5 types d'ingrédients (entrées) peuvent être reconnus:

* La tomate
* Le poivron
* Le chocolat
* Le fromage
* Le piment

Le neurone artificiel est représenté par le plateau de jeu suivant:

L'effet excitateur (ou inhibiteur) des ingrédients est représenté par le nombre de jetons verts (ou rouges) situés le long des flèches. Le corps somatique du neurone miam est représenté par le bol. Le seuil d'excitation du neurone miam est représenté par le(s) jeton(s) situé près du bol (1 jeton vert ici). </div>

Le neurone est déjà entrainé, et l'influence positive ou négative des ingrédients sur le neurone miam est donnée par le nombre et la couleur des jetons.

## Matériel pour le jeu:

* Les cinq cartes représentant les cinq ingrédients (recto avec ingrédient, et verso avec ingrédient barré d'une croix), et la carte représentant le smiley miam.
* 4 jetons vert et 5 jetons rouges.
* Un bol pour représenter le corps cellulaire.
* 6 bandes de papiers en forme de flèches, qui relient les cinq cartes vers le bol, et le bol vers la carte miam.
* Un sac. Le sac servira à mettre les cartes ingrédients, qui seront tirées au sort par le cuisinier. 

## Préparation du jeu (pour 7 élèves)

* Les cartes, bandes fléchées, bol et jetons sont disposés comme sur la figure 3.
* Un élève joue le rôle du cuisinier, 5 élèves jouent le rôle des 'neurones ingrédients', un élève le rôle du neurone miam. 

Note: On peut adapter le jeu avec plus ou moins d'élèves, soit en ajoutant/retirant des ingrédients, ou en groupant les élèves pour les différents rôles.

## Règles du jeu

* A chaque tour de jeu:
    1. L'élève jouant le rôle du neurone miam retire la carte miam et la met de côté (on ne sait pas encore si la pizza sera bonne). L'élève jouant le rôle du cuisinier retire les 5 cartes ingrédients et les place dans le sac.
    2. Le cuisinier tire au hasard entre 1 et 5 cartes ingrédients du sac.
    3. Les élèves jouant le rôle des neurones ingrédients placent les cartes ingrédients tirées côté recto (non barré). Les cartes non tirées sont sorties du sac et placées côté verso (barré): l'ingrédient n'est pas présent.
    4. Les élèves jouant le rôle des neurones ingrédients présents (non barrés) transmettent les jetons (les neurotransmetteurs) dans le bol (le corps somatique).
    5. Le neurone miam calcule la différence entre le nombre de jetons verts et rouges. Si cette différence est supérieure au seuil d'activité (le nombre de jetons verts près du bol, soit 1 ici), il pose la carte côté miam en sortie du neurone: La pizza est bonne! Sinon il pose la carte côté beurk, la pizza n'est pas bonne.
    6. Les jetons sont ensuite replacés comme sur la figure 1.

# Aller plus loin: Programmation en Python 

Cette activité avancée montre comment le neurone miam, ainsi que son apprentissage, peuvent être programmés avec le langage Python.



## Neurone miam

L'algorithme pour calculer la sortie du neurone miam est le suivant. 

```
Fonction neurone_miam:

Variables d'entrées:
* nombe_ingredients: Le nombre d'ingrédients dans la pizza (ici, 5)
* liste_neurone_ingredients: une liste des 5 ingrédients, contenant pour chaque ingrédient 1 si l'ingrédient est présent, et 0 sinon. Par exemple [1,0,0,1,0] si la tomate et le fromage sont présents.
* liste_poids_synaptiques_ingredients: une liste des 5 poids synaptiques, contenant pour chaque ingrédient le poids synaptique de l'ingrédient. Par exemple [1,1,-2,1,-3] pour le neurone miam entrainé.
* seuil_neurone_miam: Le seuil d'activation du neurone miam. Par exemple 1 pour le neurone entrainé.

Variable de sortie: 
* activation_miam: 1 si le neurone s'active, et 0 sinon.

Algorithme:

1. Calculer la somme des stimulations, et le stocker dans une variable somme_stimulations. C'est la sommme des entrées mutipliées par les poids synaptiques. On fera une boucle parcourant chaque ingrédient, et ajoutant le produit de l'entrée par le poids synaptique à la variable somme_stimulations
2. Mettre 0 dans la variable de sortie activation_miam  (le neurone ne s'active pas par défaut)
3. Si somme_activation est supérieur à seuil_neurone_miam, alors mettre 1 dans la variable de sortie activation_miam.
4. La fonction renvoie en résultat la variable de sortie activation_miam

```


La traduction de cet algorithme en langage Python donne:

````python
def neurone_miam(nombre_ingredients,
                 liste_neurone_ingredients,
                 liste_poids_synaptiques_ingredients,
                 seuil_neurone_miam):
    
    somme_activation = 0
    
    for i in range(0,nombre_ingredients):
        somme_activation = somme_activation + liste_neurone_ingredients[i]*liste_poids_synaptiques_ingredients[i]
        
    activation_miam = 0
    if somme_activation > seuil_neurone_miam:
        activation_miam = 1
    
    return activation_miam
````

Testons que le neurone s'active avec les ingrédients tomate et fromage:

````python
activation_miam = neurone_miam(nombre_ingredients=5,
                               liste_neurone_ingredients=[1,0,0,1,0],
                               liste_poids_synaptiques_ingredients=[1,1,-2,1,-3],
                               seuil_neurone_miam=1)
print("Le neurone miam a renvoyé "+str(activation_miam))
````

La sortie est à 1, le neurone est donc bien activé. On peut également tester un exemple de mauvaise pizza (poivron/chocolat):
    
    

````python
activation_miam = neurone_miam(nombre_ingredients=5,
                               liste_neurone_ingredients=[0,1,1,0,0],
                               liste_poids_synaptiques_ingredients=[1,1,-2,1,-3],
                               seuil_neurone_miam=1)
print("Le neurone miam a renvoyé "+str(activation_miam))
````

```
* liste_poids_synaptiques_ingredients: une liste des 5 poids synaptiques, contenant pour chaque ingrédient le poids synaptique de l'ingrédient. Par exemple [0,0,0,0,0] si le neurone miam n'est pas entrainé.
* seuil_neurone_miam: Le seuil d'activation du neurone miam. Par exemple 0 pour si neurone n'est pas encore entrainé.

Variable de sortie: 
* liste_poids_synaptiques_ingredients: Les poids synaptiques, éventuellement corrigés.
* seuil_neurone_miam: Le seuil d'excitiation, éventuellement corrigé.

Algorithme:

1. Calculer la sortie du neurone miam pour la pizza donnée (étapes 1 à 5 de l'activité 2), et stocker la valeur dans une variable activation_miam. On utilise la fonction neurone_miam.
2. Si l'activation est à zéro (le neurone s'est trompé), alors
    3. Corriger les poids synaptique des ingrédients présents en les augmentant de un (étape 6 de l'activité 2).
    4. Corriger le seuil d'excitabilité du neurone en le diminuant de 1 (étape 7 de l'activité 2)

```


#### La programmation


La traduction de cet algorithme en langage Python donne:

````python
def entrainement_avec_bonne_pizza(nombre_ingredients,
                                  liste_neurone_ingredients,
                                  liste_poids_synaptiques_ingredients,
                                  seuil_neurone_miam
                                 ):
    
    # On calcule l'activation du neurone miam pour les ingrédients donnés (étapes 1-5 de l'activité 2)
    activation_miam = neurone_miam(nombre_ingredients=nombre_ingredients,
                                   liste_neurone_ingredients=liste_neurone_ingredients,
                                   liste_poids_synaptiques_ingredients=liste_poids_synaptiques_ingredients,
                                   seuil_neurone_miam=seuil_neurone_miam)
        
    # Si le neurone ne s'active pas, alors il faut corriger le neurone
    if activation_miam == 0:
            
        # Pour chaque ingrédient
        for ingredient in range(nombre_ingredients):
                
            # Si l'ingrédient est présent, on augmente de 1 le poids synaptique (étape 6 de l'activité 2)
            if liste_neurone_ingredients[ingredient]==1:
                liste_poids_synaptiques_ingredients[ingredient] = liste_poids_synaptiques_ingredients[ingredient] + 1
                
        # Et on diminue de 1 le seuil d'activité (étape 7 de l'activité 2)
        seuil_neurone_miam = seuil_neurone_miam - 1
            
    # La fonction renvoie la liste des poids synaptiques et le seuil d'activité éventuellement corrigés
    return (liste_poids_synaptiques_ingredients, seuil_neurone_miam)
    
    
````

## Programme final

Nous avons maintenant tous les éléments pour écrire un programme qui reproduit l'activité 2. 

Trois fonctions ont été écrites:

* neurone_miam : Pour calculer la sortie du neurnoe miam en fonction des ingrédients
* entrainement_avec_bonne_pizza: Pour entraîner le neurone avec une exemple de bonne pizza
* entrainement_avec_mauvaise_pizza: Pour entraîner le neurone avec une exemple de mauvaise pizza

Les trois fonctions sont redonnées ci-dessous

````python
def neurone_miam(nombre_ingredients,
                 liste_neurone_ingredients,
                 liste_poids_synaptiques_ingredients,
                 seuil_neurone_miam):
    
    somme_activation = 0
    
    for i in range(0,nombre_ingredients):
        somme_activation = somme_activation + liste_neurone_ingredients[i]*liste_poids_synaptiques_ingredients[i]
        
    activation_miam = 0
    if somme_activation > seuil_neurone_miam:
        activation_miam = 1
    
    return activation_miam

def entrainement_avec_bonne_pizza(nombre_ingredients,
                                  liste_neurone_ingredients,
                                  liste_poids_synaptiques_ingredients,
                                  seuil_neurone_miam
                                 ):
    
    # On calcule l'activation du neurone miam pour les ingrédients donnés (étapes 1-5 de l'activité 2)
    activation_miam = neurone_miam(nombre_ingredients=nombre_ingredients,
                                   liste_neurone_ingredients=liste_neurone_ingredients,
                                   liste_poids_synaptiques_ingredients=liste_poids_synaptiques_ingredients,
                                   seuil_neurone_miam=seuil_neurone_miam)
        
    # Si le neurone ne s'active pas, alors il faut corriger le neurone
    if activation_miam == 0:
            
        # Pour chaque ingrédient
        for ingredient in range(nombre_ingredients):
                
            # Si l'ingrédient est présent, on augmente de 1 le poids synaptique (étape 6 de l'activité 2)
            if liste_neurone_ingredients[ingredient]==1:
                liste_poids_synaptiques_ingredients[ingredient] = liste_poids_synaptiques_ingredients[ingredient] + 1
                
        # Et on diminue de 1 le seuil d'activité (étape 7 de l'activité 2)
        seuil_neurone_miam = seuil_neurone_miam - 1
            
    # La fonction renvoie la liste des poids synaptiques et le seuil d'activité éventuellement corrigés
    return (liste_poids_synaptiques_ingredients, seuil_neurone_miam)
    
def entrainement_avec_mauvaise_pizza(nombre_ingredients,
                                     liste_neurone_ingredients,
                                     liste_poids_synaptiques_ingredients,
                                     seuil_neurone_miam
                                    ):
    
    # On calcule l'activation du neurone miam pour les ingrédients donnés (étapes 1-5 de l'activité 2)
    activation_miam = neurone_miam(nombre_ingredients=nombre_ingredients,
                                   liste_neurone_ingredients=liste_neurone_ingredients,
                                   liste_poids_synaptiques_ingredients=liste_poids_synaptiques_ingredients,
                                   seuil_neurone_miam=seuil_neurone_miam)
        
    # Si le neurone s'active, alors il faut corriger le neurone
    if activation_miam == 1:
            
        # Pour chaque ingrédient
        for ingredient in range(nombre_ingredients):
                
            # Si l'ingrédient est présent, on diminue de 1 le poids synaptique (étape 6 de l'activité 2)
            if liste_neurone_ingredients[ingredient]==1:
                liste_poids_synaptiques_ingredients[ingredient] = liste_poids_synaptiques_ingredients[ingredient] - 1
                
        # Et on augmente de 1 le seuil d'activité (étape 7 de l'activité 2)
        seuil_neurone_miam = seuil_neurone_miam + 1
            
    # La fonction renvoie la liste des poids synaptiques et le seuil d'activité éventuellement corrigés
    return (liste_poids_synaptiques_ingredients, seuil_neurone_miam)
      
    
    
````
